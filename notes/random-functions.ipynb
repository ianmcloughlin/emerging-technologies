## Select One At Random

```python
[f1, f2, f3, f4]
```

```python
def random_function():
  # List of potential functions.
  fns = [f1, f2, f3, f4]
  # Return element at random.
  return random.choice(fns)
  
```

```python
f = random_function()
```

```python
f(0)
```

```python
f(1)
```

```python
f
```

```python
def random_function():
  f1 = lambda a: 0
  f2 = lambda a: a
  f3 = lambda a: 0 if a else 1
  f4 = lambda a: 1
  
  fns = [f1, f2, f3, f4]
  # Return element at random.
  return random.choice(fns)
```

```python
f = random_function()
```

```python
f
```

```python
f(0), f(1)
```
